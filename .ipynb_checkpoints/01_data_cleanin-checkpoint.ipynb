### Step 4- Data Cleaning

**1 - HR Dataset (Employee Attrition)**

```python
import pandas as pd

# Load HR dataset
hr = pd.read_csv("data/raw/HR-Employee-Attrition.csv")

# Drop duplicates
hr_clean = hr.drop_duplicates().copy()

# Convert Attrition Yes/No → 1/0
if "Attrition" in hr_clean.columns:
    hr_clean["AttritionFlag"] = hr_clean["Attrition"].map({"Yes": 1, "No": 0})

# Convert other Yes/No columns automatically
for col in hr_clean.columns:
    if hr_clean[col].dtype == "object" and hr_clean[col].nunique() == 2:
        hr_clean[col] = hr_clean[col].map({"Yes": 1, "No": 0})

# Save cleaned HR dataset
hr_clean.to_csv("data/cleaned/hr_clean.csv", index=False)

print("HR: Before cleaning:", hr.shape, "→ After cleaning:", hr_clean.shape)
```

**2.Superstore Dataset (Sales)**

```python
# Load Superstore dataset
superstore = pd.read_excel("data/raw/Sample - Superstore.xls")

# Drop duplicates
superstore_clean = superstore.drop_duplicates().copy()

# Drop rows with missing postal codes
superstore_clean = superstore_clean.dropna(subset=["Postal Code"])

# Save cleaned Superstore dataset
superstore_clean.to_csv("data/cleaned/superstore_clean.csv", index=False)

print("Superstore: Before cleaning:", superstore.shape, "→ After cleaning:", superstore_clean.shape)
```

**3- Online Retail Dataset (Transactions)**

```python
# Load Online Retail dataset
retail = pd.read_excel("data/raw/Online Retail.xlsx")

# Drop rows with missing CustomerID
retail_clean = retail.dropna(subset=["CustomerID"]).copy()

# Fill missing product descriptions with "Unknown"
retail_clean["Description"] = retail_clean["Description"].fillna("Unknown")

# Remove negative or zero Quantity (invalid transactions/returns)
retail_clean = retail_clean[retail_clean["Quantity"] > 0]

# Remove negative or zero UnitPrice
retail_clean = retail_clean[retail_clean["UnitPrice"] > 0]

# Convert InvoiceDate to datetime format
retail_clean["InvoiceDate"] = pd.to_datetime(retail_clean["InvoiceDate"], errors="coerce")

# Drop duplicates
retail_clean = retail_clean.drop_duplicates()

# Save cleaned Retail dataset
retail_clean.to_csv("data/cleaned/retail_clean.csv", index=False)

print("Retail: Before cleaning:", retail.shape, "→ After cleaning:", retail_clean.shape)
```

```python
import pandas as pd

hr = pd.read_csv("data/cleaned/hr_clean.csv")
superstore = pd.read_csv("data/cleaned/superstore_clean.csv")
retail = pd.read_csv("data/cleaned/retail_clean.csv")

print("HR shape:", hr.shape)
print("Superstore shape:", superstore.shape)
print("Retail shape:", retail.shape)

print("Missing values check:")
print(hr.isnull().sum().sum(), superstore.isnull().sum().sum(), retail.isnull().sum().sum())
```
