## 1 - imports, folders

```python
import os
from pathlib import Path
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

# Create folders for outputs
os.makedirs("data/analysis", exist_ok=True)
os.makedirs("docs/images", exist_ok=True)

# plotting default
plt.rcParams.update({'figure.figsize': (10,5), 'figure.dpi': 120, 'font.size': 12})
```

## 2 - Load cleaned datasets & normalize column names

```python

def normalize_cols(df):
    df = df.copy()
    df.columns = (
        df.columns.str.strip()
                  .str.lower()
                  .str.replace(" ", "_")
                  .str.replace(r"[^\w_]", "", regex=True)
    )
    return df

# load
hr = normalize_cols(pd.read_csv("data/cleaned/hr_clean.csv"))
superstore = normalize_cols(pd.read_csv("data/cleaned/superstore_clean.csv"))
retail = normalize_cols(pd.read_csv("data/cleaned/retail_clean.csv"))

# parse common date columns
for df,name in [(superstore,'superstore'), (retail,'retail')]:
    # try common names
    for cand in ['order_date','orderdate','invoice_date','invoicedate','date']:
        if cand in df.columns:
            df[cand] = pd.to_datetime(df[cand], errors='coerce')
            # save canonical name
            if name=='superstore': superstore['order_date'] = df[cand]
            if name=='retail': retail['invoice_date'] = df[cand]
            break

# check
print("HR shape:", hr.shape)
print("Superstore shape:", superstore.shape, "date col present:", 'order_date' in superstore.columns)
print("Retail shape:", retail.shape, "date col present:", 'invoice_date' in retail.columns)
```

## 3 - Ensure revenue fields exist (compute retail revenue)

```python
# Superstore: try to find sales column
if 'sales' in superstore.columns:
    superstore['revenue'] = pd.to_numeric(superstore['sales'], errors='coerce').fillna(0)
else:
    raise SystemExit("Superstore: no 'sales' column found. Check column names: \n" + str(superstore.columns.tolist()))

# Retail: compute revenue from quantity * unitprice
qty_col = next((c for c in retail.columns if 'quantity' in c), None)
price_col = next((c for c in retail.columns if 'unitprice' in c or 'unit_price' in c or 'price' in c), None)
if qty_col and price_col:
    retail['quantity'] = pd.to_numeric(retail[qty_col], errors='coerce').fillna(0)
    retail['unit_price'] = pd.to_numeric(retail[price_col], errors='coerce').fillna(0)
    retail['revenue'] = retail['quantity'] * retail['unit_price']
else:
    # if already has revenue column
    if 'revenue' not in retail.columns:
        raise SystemExit("Retail: could not find quantity/unitprice columns. Check names: \n" + str(retail.columns.tolist()))

# HR: ensure attrition flag present (0/1)
if 'attritionflag' not in hr.columns and 'attrition' in hr.columns:
    hr['attritionflag'] = hr['attrition'].map(lambda x: 1 if str(x).strip().lower() in ['yes','y','1', 'true'] else 0)

print("Revenue columns prepared. Examples:")
print(" Superstore revenue head:\n", superstore['revenue'].head())
print(" Retail revenue head:\n", retail['revenue'].head())
print(" HR attrition unique:", hr['attritionflag'].unique()[:5])
```

## 3 - KPI

```python
# Single robust cell to inspect columns, auto-detect fields, compute simple KPIs, save results.
import re, os, json
import pandas as pd
import numpy as np
from pathlib import Path

# -- 1) (re)load dataframes 

if 'superstore' not in globals():
    superstore = pd.read_csv("data/cleaned/superstore_clean.csv")
if 'retail' not in globals():
    retail = pd.read_csv("data/cleaned/retail_clean.csv")
if 'hr' not in globals():
    hr = pd.read_csv("data/cleaned/hr_clean.csv")

# helper: normalized column name

def norm(col): 
    return re.sub(r'[^a-z0-9]', '', str(col).lower())

def find_col(df, keywords):
    """Return first column in df whose normalized name contains any keyword string."""
    norm_map = {norm(c): c for c in df.columns}
    for kw in keywords:
        for n, orig in norm_map.items():
            if kw in n:
                return orig
    return None

def list_numeric_candidates(df):
    return [c for c in df.columns if pd.api.types.is_numeric_dtype(df[c])]

# show columns + sample

print("=== Superstore columns ===")
print(list(superstore.columns))
print("\nSample rows (superstore):")
display(superstore.head(3))

print("\n=== Retail columns ===")
print(list(retail.columns))
print("\nSample rows (retail):")
display(retail.head(3))

print("\n=== HR columns ===")
print(list(hr.columns))
print("\nSample rows (hr):")
display(hr.head(3))

# -- 2) Auto-detect columns 

sales_col = find_col(superstore, ['sales','revenue','amount','total'])
order_col = find_col(superstore, ['orderid','order_id','orderid','orderno','order'])

# Retail: quantity, price, invoice, customer

qty_col = find_col(retail, ['quantity','qty','qtysold'])
price_col = find_col(retail, ['unitprice','unit_price','price','unitprice'])
retail_revenue_col = find_col(retail, ['revenue','amount','total','sales'])
invoice_col = find_col(retail, ['invoiceno','invoice_no','invoice','invoiceid','invoicenumber'])
cust_col = find_col(retail, ['customerid','customer_id','customer','custid'])

# Retail date column

retail_date_col = find_col(retail, ['invoicedate','invoice_date','date','orderdate','dateinvoice'])
super_date = find_col(superstore, ['orderdate','order_date','date'])

# HR attrition column

hr_attr_col = find_col(hr, ['attritionflag','attrition','left','is_left'])

# Feedback to user

print("\nDetected (best guess) columns:")
print(" superstore.sales_col ->", sales_col)
print(" superstore.order_col ->", order_col)
print(" superstore.date_col  ->", super_date)
print(" retail.revenue_col   ->", retail_revenue_col)
print(" retail.qty_col       ->", qty_col)
print(" retail.price_col     ->", price_col)
print(" retail.invoice_col   ->", invoice_col)
print(" retail.customer_col  ->", cust_col)
print(" retail.date_col      ->", retail_date_col)
print(" hr.attrition_col     ->", hr_attr_col)

# -- 3) Compute KPIs (use fallbacks, safe conversions)

# Ensure output folder

os.makedirs("data/analysis", exist_ok=True)

# Superstore revenue

if sales_col is None:
    # try numeric candidates as fallback (pick the largest-sum numeric column as a guess)
    
    nums = list_numeric_candidates(superstore)
    if nums:
        totals = {c: pd.to_numeric(superstore[c], errors='coerce').fillna(0).sum() for c in nums}
        # choose column with largest sum (likely revenue)
        sales_col = max(totals, key=totals.get)
        print(f"\n[Fallback] Using numeric column '{sales_col}' as Sales (largest numeric sum).")
    else:
        raise Exception("Could not detect a Sales-like column in superstore. Inspect columns above.")

superstore_revenue = pd.to_numeric(superstore[sales_col], errors='coerce').fillna(0).sum()

# AOV (Average Order Value) - try to find order id

if order_col is None:
    # try invoice/order-like columns
    possible = [c for c in superstore.columns if 'order' in norm(c) or 'invoice' in norm(c)]
    order_col = possible[0] if possible else None
    if order_col:
        print(f"[Fallback] Using '{order_col}' as order identifier.")
        
if order_col:
    try:
        unique_orders = superstore[order_col].nunique()
        aov = superstore_revenue / unique_orders if unique_orders else superstore_revenue / len(superstore)
    except Exception:
        aov = superstore_revenue / len(superstore)
else:
    aov = superstore_revenue / len(superstore)

# Retail revenue

if retail_revenue_col is not None:
    retail_revenue = pd.to_numeric(retail[retail_revenue_col], errors='coerce').fillna(0).sum()
else:
    # try compute from qty * price
    if qty_col and price_col:
        q = pd.to_numeric(retail[qty_col], errors='coerce').fillna(0)
        p = pd.to_numeric(retail[price_col], errors='coerce').fillna(0)
        retail['__computed_revenue__'] = q * p
        retail_revenue = retail['__computed_revenue__'].sum()
        print(f"[Computed] Retail revenue = {qty_col} * {price_col}")
    else:
        # fallback: pick largest numeric column
        nums = list_numeric_candidates(retail)
        if nums:
            totals = {c: pd.to_numeric(retail[c], errors='coerce').fillna(0).sum() for c in nums}
            retail_revenue = totals[max(totals, key=totals.get)]
            print(f"[Fallback] Using numeric column '{max(totals, key=totals.get)}' as Retail revenue.")
        else:
            raise Exception("Could not detect retail revenue. Check quantity/unit price columns.")

# HR attrition rate: find & compute

attrition_rate = None
if hr_attr_col:
    # try numeric
    if pd.api.types.is_numeric_dtype(hr[hr_attr_col]):
        attrition_rate = hr[hr_attr_col].mean()
    else:
        # map yes/no to 1/0
        mapped = hr[hr_attr_col].astype(str).str.strip().str.lower().map(lambda x: 1 if x in ('yes','y','1','true') else 0)
        attrition_rate = mapped.mean()
else:
    # try to find a text column 'attrition' or 'status' 
    alt = find_col(hr, ['attrition','left','separated','turnover'])
    if alt:
        mapped = hr[alt].astype(str).str.strip().str.lower().map(lambda x: 1 if x in ('yes','y','1','true') else 0)
        attrition_rate = mapped.mean()
        print(f"[Fallback] Using '{alt}' for attrition.")
    else:
        print("Warning: Could not detect attrition column in HR dataset.")
        attrition_rate = np.nan

# churn rate (retail): simple definition as customers with last_purchase older than 3 months from max

churn_rate = None
# try detect invoice date and customer id columns
retail_date = retail_date_col or find_col(retail, ['date','orderdate','invoicedate'])
cust_col = cust_col or find_col(retail, ['customerid','customer','cust'])
invoice_col = invoice_col or find_col(retail, ['invoice','invoice_no','invoiceno','invoiceno'])

if retail_date and cust_col and invoice_col:
    # parse date
    retail[retail_date] = pd.to_datetime(retail[retail_date], errors='coerce')
    cs = retail.groupby(cust_col).agg(last_purchase=(retail_date, 'max')).reset_index()
    if cs['last_purchase'].notnull().any():
        cutoff = cs['last_purchase'].max() - pd.DateOffset(months=3)
        churn_count = (cs['last_purchase'] < cutoff).sum()
        churn_rate = churn_count / len(cs) if len(cs)>0 else np.nan
    else:
        churn_rate = np.nan
else:
    # not enough data to compute churn
    churn_rate = np.nan
    print("Note: Could not compute churn; missing invoice date or customer id or invoice columns in retail.")

# -- 4) Prepare kpis dict & save

kpis = {
    "superstore_revenue": float(superstore_revenue),
    "retail_revenue": float(retail_revenue),
    "total_revenue": float(superstore_revenue + retail_revenue),
    "aov_superstore": float(aov),
    "employee_attrition_rate": float(attrition_rate) if not pd.isna(attrition_rate) else None,
    "customer_churn_rate": float(churn_rate) if not pd.isna(churn_rate) else None,
    "detected_columns": {
        "superstore_sales_col": sales_col,
        "superstore_order_col": order_col,
        "retail_qty_col": qty_col,
        "retail_price_col": price_col,
        "retail_revenue_col": retail_revenue_col,
        "retail_invoice_col": invoice_col,
        "retail_customer_col": cust_col,
        "retail_date_col": retail_date,
        "hr_attrition_col": hr_attr_col
    }
}

# save json + csv

with open("data/analysis/kpis_simple.json", "w") as f:
    json.dump(kpis, f, indent=2)
pd.DataFrame([kpis]).to_csv("data/analysis/kpis_simple.csv", index=False)

print("\n=== KPIs computed ===")
print(f"Superstore revenue: {kpis['superstore_revenue']:,}")
print(f"Retail revenue:     {kpis['retail_revenue']:,}")
print(f"Total revenue:      {kpis['total_revenue']:,}")
print(f"AOV (Superstore):   {kpis['aov_superstore']:.2f}")
print(f"Employee attrition rate: {kpis['employee_attrition_rate'] if kpis['employee_attrition_rate'] is not None else 'N/A'}")
print(f"Customer churn rate:     {kpis['customer_churn_rate'] if kpis['customer_churn_rate'] is not None else 'N/A'}")
print("\nSaved to data/analysis/kpis_simple.json and .csv")
print("\nIf any KPI looks wrong, check which columns were detected and adjust column names or rename columns in the CSVs.")
```

## 4 - Charts

**4.1 Monthly Revenue Trend (Superstore)**

```python
superstore["order date"] = pd.to_datetime(superstore["order_date"])
monthly_revenue = superstore.groupby(superstore["order_date"].dt.to_period("M"))["sales"].sum()

plt.figure(figsize=(10,5))
monthly_revenue.plot(kind="line", marker="o")
plt.title("Monthly Revenue - Superstore")
plt.ylabel("Revenue")
plt.xlabel("Month")
plt.xticks(rotation=45)
plt.show()
plt.savefig("docs/images/monthly_revenue_combined.png")
```

## 4.2 Top 10 products

```python
 — top products
# Superstore product column
prod_col_sup = next((c for c in superstore.columns if 'product' in c and ('name' in c or 'title' in c)), 'product_name')
top_sup = (superstore.groupby(prod_col_sup)['revenue'].sum()
                      .sort_values(ascending=False).head(10).reset_index())

top_sup.to_csv("data/analysis/top_products_superstore.csv", index=False)

# Plot horizontal bar
fig, ax = plt.subplots(figsize=(10,5))
ax.barh(top_sup[prod_col_sup][::-1], top_sup['revenue'][::-1])
ax.set_title("Top 10 Products — Superstore (by Revenue)")
ax.set_xlabel("Revenue")
ax.xaxis.set_major_formatter(mtick.StrMethodFormatter('${x:,.0f}'))
plt.tight_layout()
plt.savefig("docs/images/top_products_superstore.png")
plt.show()


# Retail: description prod
desc_col = next((c for c in retail.columns if 'desc' in c or 'description' in c), None)
if desc_col:
    top_ret = (retail.groupby(desc_col)['revenue'].sum()
                   .sort_values(ascending=False).head(10).reset_index())
    top_ret.to_csv("data/analysis/top_products_retail.csv", index=False)
    fig, ax = plt.subplots(figsize=(10,5))
    ax.barh(top_ret[desc_col][::-1], top_ret['revenue'][::-1])
    ax.set_title("Top 10 Items — Retail (by Revenue)")
    ax.set_xlabel("Revenue")
    ax.xaxis.set_major_formatter(mtick.StrMethodFormatter('${x:,.0f}'))
    plt.tight_layout()
    plt.savefig("docs/images/top_products_retail.png")
    plt.show()
else:
    print("Retail description column not found.")
```

### 4.3 - Revenue by region(superstore)

```python

region_col = next((c for c in superstore.columns if 'region' in c), 'region')
rev_region = superstore.groupby(region_col)['revenue'].sum().sort_values(ascending=False).reset_index()
rev_region.to_csv("data/analysis/revenue_by_region.csv", index=False)

fig, ax = plt.subplots(figsize=(8,4))
ax.bar(rev_region[region_col], rev_region['revenue'])
ax.set_title("Revenue by Region — Superstore")
ax.set_ylabel("Revenue")
ax.set_xlabel("Region")
ax.yaxis.set_major_formatter(mtick.StrMethodFormatter('${x:,.0f}'))
plt.xticks(rotation=30)
plt.tight_layout()
plt.savefig("docs/images/revenue_by_region.png")
plt.show()
```

### 4.4 customer summary, repeat rate

```python

cust_col = next((c for c in retail.columns if 'customer' in c), 'customerid')
invoice_col = next((c for c in retail.columns if 'invoice' in c), 'invoiceno')

cust_summary = (retail.groupby(cust_col).agg(
    orders = (invoice_col, pd.Series.nunique),
    total_spent = ('revenue','sum'),
    last_purchase = ('invoice_date','max')
).reset_index().sort_values('total_spent', ascending=False))

cust_summary.to_csv("data/analysis/customer_summary_retail.csv", index=False)

repeat_rate = (cust_summary['orders'] > 1).mean()
print(f"Retail Repeat Purchase Rate: {repeat_rate:.2%}")

# distribution of orders (histogram)
fig, ax = plt.subplots(figsize=(8,4))
ax.hist(cust_summary['orders'], bins=20)
ax.set_title("Distribution: Orders per Customer (Retail)")
ax.set_xlabel("Number of Orders"); ax.set_ylabel("Number of Customers")
plt.tight_layout()
plt.savefig("docs/images/customer_orders_distribution.png")
plt.show()
```

## 5 - Attrition by Department (HR)

```python
if 'department' in hr.columns and 'attritionflag' in hr.columns:
    attr_by_dept = hr.groupby('department')['attritionflag'].mean().sort_values(ascending=False).reset_index()
    attr_by_dept.to_csv("data/analysis/attrition_by_dept.csv", index=False)

    fig, ax = plt.subplots(figsize=(8,4))
    ax.barh(attr_by_dept['department'][::-1], attr_by_dept['attritionflag'][::-1])
    ax.set_title("Attrition Rate by Department")
    ax.set_xlabel("Attrition Rate")
    ax.xaxis.set_major_formatter(mtick.PercentFormatter(1.0))
    plt.tight_layout()
    plt.savefig("docs/images/attrition_by_dept.png")
    plt.show()
else:
    print("HR: department or attritionflag missing")
```

### Job satisfaction 

```python
satis = next((c for c in hr.columns if 'satisf' in c or 'jobinvolvement' in c or 'job_satisfaction' in c), None)
if satis:
    fig, ax = plt.subplots(figsize=(8,4))
    ax.hist(hr[satis].dropna(), bins=6)
    ax.set_title(f"Distribution of {satis}")
    plt.tight_layout()
    plt.savefig(f"docs/images/{satis}_distribution.png")
    plt.show()
```
